For a given value of r, we can estimate the stable map iterates by applying the map for nsteps, after discarding the transient (nskip steps). In the following we show the process for a period-4 cycle.

```python
r=0.87
x=0.001
nskip=10000
for i in range(nskip): x=4*r*x*(1-x)
nsteps=100
stored_x = [x]
for i in range(nsteps):
    x=4*r*x*(1-x)
    stored_x.append(x)
plt.plot(stored_x)
plt.show()
```

From the above iterates we can compute the power spectrum of the map using fast Fourier transforms. 

```python
spectrum=np.fft.fft(stored_x)
plt.plot(spectrum.astype('float')) # the spectrum results are complex number and we only need the real part
plt.yscale('log')
plt.show()
```
